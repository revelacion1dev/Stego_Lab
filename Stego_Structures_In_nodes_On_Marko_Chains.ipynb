```python
# üì¶ Instalaci√≥n simple
!pip install hmmlearn networkx matplotlib numpy --quiet

import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from hmmlearn import hmm
import warnings
warnings.filterwarnings('ignore')

class SimpleMarkovHiding:
    """Sistema simple para ocultar informaci√≥n en cadenas de Markov"""

    def __init__(self, original_transition_matrix, n_vocab=6):
        self.original = np.array(original_transition_matrix)
        self.n_states = len(self.original)
        self.n_vocab = n_vocab
        self.hidden_matrix = None
        self.learned_matrix = None
        self.key = None

        # Validar y normalizar
        self._normalize_matrix()

        # Crear emisiones simples
        np.random.seed(42)
        self.emissions = np.random.dirichlet(np.ones(self.n_vocab), size=self.n_states)

    def _normalize_matrix(self):
        """Normalizar matriz para que sea estoc√°stica"""
        row_sums = self.original.sum(axis=1, keepdims=True)
        row_sums[row_sums == 0] = 1
        self.original = self.original / row_sums
        print(f"‚úÖ Matriz normalizada: {self.n_states} estados")

    def method_1_node_shuffle(self, seed=42):
        """M√âTODO 1: Reordenar nodos (permutaci√≥n)"""
        print("üîÄ Aplicando permutaci√≥n de nodos...")

        np.random.seed(seed)
        permutation = np.random.permutation(self.n_states)

        # Aplicar permutaci√≥n
        self.hidden_matrix = self.original[permutation][:, permutation]
        self.key = {'method': 'shuffle', 'permutation': permutation}

        return self.hidden_matrix

    def method_2_noise_injection(self, noise_level=0.1):
        """M√âTODO 2: Inyectar ruido controlado"""
        print(f"üéõÔ∏è Inyectando ruido (nivel: {noise_level})...")

        # A√±adir ruido gaussiano
        noise = np.random.normal(0, noise_level, self.original.shape)
        noisy_matrix = self.original + noise

        # Asegurar valores positivos y normalizar
        noisy_matrix = np.abs(noisy_matrix)
        noisy_matrix = noisy_matrix / noisy_matrix.sum(axis=1, keepdims=True)

        self.hidden_matrix = noisy_matrix
        self.key = {'method': 'noise', 'noise_level': noise_level}

        return self.hidden_matrix

    def method_3_uniform_mixing(self, mix_ratio=0.3):
        """M√âTODO 3: Mezclar con matriz uniforme"""
        print(f"üåä Mezclando con matriz uniforme (ratio: {mix_ratio})...")

        # Crear matriz uniforme
        uniform_matrix = np.ones((self.n_states, self.n_states)) / self.n_states

        # Mezclar
        mixed_matrix = (1 - mix_ratio) * self.original + mix_ratio * uniform_matrix

        self.hidden_matrix = mixed_matrix
        self.key = {'method': 'uniform_mix', 'mix_ratio': mix_ratio}

        return self.hidden_matrix

    def method_4_state_duplication(self):
        """M√âTODO 4: Duplicar estados para confundir"""
        print("üë• Duplicando estados...")

        # Duplicar el primer estado
        n_new = self.n_states + 1
        expanded_matrix = np.zeros((n_new, n_new))

        # Copiar matriz original
        expanded_matrix[:self.n_states, :self.n_states] = self.original

        # Duplicar primer estado en la nueva posici√≥n
        expanded_matrix[self.n_states, :self.n_states] = self.original[0, :]
        expanded_matrix[:self.n_states, self.n_states] = self.original[:, 0]
        expanded_matrix[self.n_states, self.n_states] = self.original[0, 0]

        # Normalizar
        expanded_matrix = expanded_matrix / expanded_matrix.sum(axis=1, keepdims=True)

        self.hidden_matrix = expanded_matrix
        self.key = {'method': 'duplication', 'duplicated_state': 0}

        return self.hidden_matrix

    def method_5_cycle_shift(self, shift=1):
        """M√âTODO 5: Desplazamiento c√≠clico"""
        print(f"üîÑ Aplicando desplazamiento c√≠clico (shift: {shift})...")

        # Crear matriz de desplazamiento c√≠clico
        shifted_matrix = np.zeros_like(self.original)

        for i in range(self.n_states):
            for j in range(self.n_states):
                new_i = (i + shift) % self.n_states
                new_j = (j + shift) % self.n_states
                shifted_matrix[new_i, new_j] = self.original[i, j]

        self.hidden_matrix = shifted_matrix
        self.key = {'method': 'cycle_shift', 'shift': shift}

        return self.hidden_matrix

    def generate_sequences(self, matrix, n_sequences=200, seq_length=30):
        """Generar secuencias desde una matriz de transici√≥n"""
        sequences = []
        states_list = []

        for _ in range(n_sequences):
            # Estado inicial aleatorio
            current_state = np.random.choice(matrix.shape[0])

            sequence = []
            states = []

            for _ in range(seq_length):
                states.append(current_state)

                # Generar observaci√≥n
                obs = np.random.choice(self.n_vocab, p=self.emissions[current_state % self.n_states])
                sequence.append([obs])

                # Transici√≥n
                current_state = np.random.choice(matrix.shape[0], p=matrix[current_state])

            sequences.append(np.array(sequence))
            states_list.append(np.array(states))

        return sequences, states_list

    def train_hmm(self, sequences):
        """Entrenar HMM con las secuencias"""
        print("üß† Entrenando HMM...")

        # Preparar datos
        X_concat = np.concatenate(sequences)
        lengths = [len(seq) for seq in sequences]

        # Entrenar HMM
        model = hmm.MultinomialHMM(n_components=self.n_states, n_iter=100, random_state=42)
        model.fit(X_concat, lengths)

        self.learned_matrix = model.transmat_

        print(f"‚úÖ HMM entrenado")
        return model

    def recover_original(self):
        """Intentar recuperar matriz original usando la clave"""
        if self.key is None or self.learned_matrix is None:
            print("‚ùå No hay clave o matriz aprendida")
            return self.learned_matrix

        method = self.key['method']

        if method == 'shuffle':
            # Aplicar permutaci√≥n inversa
            perm = self.key['permutation']
            inv_perm = np.argsort(perm)
            recovered = self.learned_matrix[inv_perm][:, inv_perm]

        elif method == 'noise':
            # Para ruido, la matriz aprendida ya deber√≠a estar cerca
            recovered = self.learned_matrix

        elif method == 'uniform_mix':
            # Intentar "desmezclar" (aproximaci√≥n)
            mix_ratio = self.key['mix_ratio']
            uniform_matrix = np.ones((self.n_states, self.n_states)) / self.n_states
            recovered = (self.learned_matrix - mix_ratio * uniform_matrix) / (1 - mix_ratio)
            recovered = np.abs(recovered)
            recovered = recovered / recovered.sum(axis=1, keepdims=True)

        else:
            recovered = self.learned_matrix

        return recovered

    def visualize_comparison(self, title="Comparaci√≥n de Grafos"):
        """Visualizar los tres grafos: original, oculto, aprendido"""

        matrices = [
            (self.original, "üéØ Grafo Original"),
            (self.hidden_matrix, "üé≠ Grafo Oculto"),
            (self.learned_matrix, "üß† Grafo Aprendido")
        ]

        # A√±adir recuperado si existe
        if hasattr(self, 'recovered_matrix'):
            matrices.append((self.recovered_matrix, "üîÑ Grafo Recuperado"))

        n_plots = len(matrices)
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))
        axes = axes.flatten()

        for idx, (matrix, label) in enumerate(matrices):
            if idx >= 4:  # M√°ximo 4 subplots
                break

            ax = axes[idx]

            # Heatmap
            im = ax.imshow(matrix, cmap='Blues', vmin=0, vmax=1)
            ax.set_title(label, fontsize=12, fontweight='bold')
            ax.set_xlabel("Estado Destino")
            ax.set_ylabel("Estado Origen")

            # A√±adir valores en las celdas
            for i in range(matrix.shape[0]):
                for j in range(matrix.shape[1]):
                    text = ax.text(j, i, f'{matrix[i, j]:.2f}',
                                 ha="center", va="center", color="black", fontsize=8)

            # Colorbar
            plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

        # Ocultar subplots no usados
        for idx in range(n_plots, 4):
            axes[idx].set_visible(False)

        plt.suptitle(title, fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()

    def visualize_networks(self):
        """Visualizar como grafos de red"""

        matrices = [
            (self.original, "üéØ Original"),
            (self.hidden_matrix, "üé≠ Oculto"),
            (self.learned_matrix, "üß† Aprendido")
        ]

        fig, axes = plt.subplots(1, 3, figsize=(15, 5))

        for idx, (matrix, title) in enumerate(matrices):
            ax = axes[idx]

            # Crear grafo
            G = nx.DiGraph()

            # A√±adir nodos y aristas
            for i in range(matrix.shape[0]):
                for j in range(matrix.shape[1]):
                    if matrix[i, j] > 0.05:  # Solo mostrar conexiones significativas
                        G.add_edge(i, j, weight=matrix[i, j])

            # Layout
            pos = nx.spring_layout(G, seed=42)

            # Dibujar nodos
            nx.draw_networkx_nodes(G, pos, node_color='lightblue',
                                 node_size=800, ax=ax)

            # Dibujar aristas con grosor proporcional al peso
            edges = G.edges()
            weights = [G[u][v]['weight'] for u, v in edges]
            nx.draw_networkx_edges(G, pos, width=[w*5 for w in weights],
                                 edge_color='gray', arrows=True, ax=ax)

            # Etiquetas de nodos
            nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold', ax=ax)

            # Etiquetas de aristas (pesos)
            edge_labels = {(u, v): f'{w:.2f}' for u, v, w in G.edges(data='weight')}
            nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=8, ax=ax)

            ax.set_title(title, fontsize=12, fontweight='bold')
            ax.axis('off')

        plt.tight_layout()
        plt.show()

    def calculate_similarity(self, matrix1, matrix2):
        """Calcular similitud entre dos matrices"""
        # Correlaci√≥n
        corr = np.corrcoef(matrix1.flatten(), matrix2.flatten())[0, 1]

        # Error Frobenius
        frobenius_error = np.linalg.norm(matrix1 - matrix2, 'fro')

        # Error m√°ximo por entrada
        max_error = np.max(np.abs(matrix1 - matrix2))

        return {
            'correlation': corr,
            'frobenius_error': frobenius_error,
            'max_error': max_error
        }

    def analyze_hiding_effectiveness(self):
        """Analizar qu√© tan efectivo fue el m√©todo de ocultaci√≥n"""
        print("\nüìä AN√ÅLISIS DE EFECTIVIDAD DE OCULTACI√ìN")
        print("=" * 50)

        # Similitud Original vs Oculto (menor = mejor ocultaci√≥n)
        hide_sim = self.calculate_similarity(self.original, self.hidden_matrix)
        print(f"üé≠ Original vs Oculto:")
        print(f"   Correlaci√≥n: {hide_sim['correlation']:.4f} (menor = mejor)")
        print(f"   Error Frobenius: {hide_sim['frobenius_error']:.4f} (mayor = mejor)")

        # Similitud Original vs Aprendido (mayor = peor ocultaci√≥n)
        learn_sim = self.calculate_similarity(self.original, self.learned_matrix)
        print(f"\nüß† Original vs Aprendido:")
        print(f"   Correlaci√≥n: {learn_sim['correlation']:.4f}")
        print(f"   Error Frobenius: {learn_sim['frobenius_error']:.4f}")

        # Similitud Oculto vs Aprendido
        hidden_sim = self.calculate_similarity(self.hidden_matrix, self.learned_matrix)
        print(f"\nüîç Oculto vs Aprendido:")
        print(f"   Correlaci√≥n: {hidden_sim['correlation']:.4f}")
        print(f"   Error Frobenius: {hidden_sim['frobenius_error']:.4f}")

        # Score de ocultaci√≥n
        hiding_score = 1 - hide_sim['correlation']  # 1 = perfecta ocultaci√≥n
        detection_risk = learn_sim['correlation']   # 0 = no detectable

        print(f"\nüèÜ SCORES FINALES:")
        print(f"   üõ°Ô∏è Score de Ocultaci√≥n: {hiding_score:.4f} (0-1, mayor = mejor)")
        print(f"   ‚ö†Ô∏è Riesgo de Detecci√≥n: {detection_risk:.4f} (0-1, menor = mejor)")

        return {
            'hiding_score': hiding_score,
            'detection_risk': detection_risk,
            'similarities': {
                'original_vs_hidden': hide_sim,
                'original_vs_learned': learn_sim,
                'hidden_vs_learned': hidden_sim
            }
        }

def demo_all_methods():
    """Demostrar todos los m√©todos de ocultaci√≥n"""
    print("üé≠ DEMO: M√âTODOS DE OCULTACI√ìN EN CADENAS DE MARKOV")
    print("=" * 60)

    # Crear matriz original interesante
    original = np.array([
        [0.1, 0.7, 0.2],
        [0.4, 0.2, 0.4],
        [0.3, 0.3, 0.4]
    ])

    print("üéØ Matriz Original:")
    print(original)

    methods = [
        ('node_shuffle', 'method_1_node_shuffle', {}),
        ('noise_injection', 'method_2_noise_injection', {'noise_level': 0.15}),
        ('uniform_mixing', 'method_3_uniform_mixing', {'mix_ratio': 0.4}),
        ('cycle_shift', 'method_5_cycle_shift', {'shift': 1})
    ]

    results = {}

    for method_name, method_func, params in methods:
        print(f"\nüî¨ PROBANDO: {method_name.upper()}")
        print("-" * 40)

        # Crear instancia del sistema
        system = SimpleMarkovHiding(original)

        # Aplicar m√©todo de ocultaci√≥n
        getattr(system, method_func)(**params)

        # Generar datos desde la matriz oculta
        sequences, _ = system.generate_sequences(system.hidden_matrix)

        # Entrenar HMM
        model = system.train_hmm(sequences)

        # Intentar recuperar
        system.recovered_matrix = system.recover_original()

        # Analizar efectividad
        analysis = system.analyze_hiding_effectiveness()

        # Visualizar
        system.visualize_comparison(f"M√©todo: {method_name}")
        system.visualize_networks()

        results[method_name] = {
            'system': system,
            'analysis': analysis
        }

    # Comparaci√≥n final
    print(f"\nüèÜ COMPARACI√ìN FINAL DE M√âTODOS")
    print("=" * 50)

    for method_name, result in results.items():
        analysis = result['analysis']
        print(f"{method_name:15} | Ocultaci√≥n: {analysis['hiding_score']:.3f} | " +
              f"Detecci√≥n: {analysis['detection_risk']:.3f}")

    # Encontrar mejor m√©todo
    best_method = max(results.keys(),
                     key=lambda x: results[x]['analysis']['hiding_score'] -
                                  results[x]['analysis']['detection_risk'])

    print(f"\nü•á MEJOR M√âTODO: {best_method}")

    return results

# üöÄ EJECUTAR DEMO
if __name__ == "__main__":
    results = demo_all_methods()
```

```python
# üì¶ Instalaci√≥n simple
!pip install hmmlearn networkx matplotlib numpy --quiet

import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from hmmlearn import hmm
import warnings
warnings.filterwarnings('ignore')

class SimpleMarkovHiding:
    """Sistema simple para ocultar informaci√≥n en cadenas de Markov"""

    def __init__(self, original_transition_matrix, n_vocab=6):
        self.original = np.array(original_transition_matrix)
        self.n_states = len(self.original)
        self.n_vocab = n_vocab
        self.hidden_matrix = None
        self.learned_matrix = None
        self.key = None

        # Validar y normalizar
        self._normalize_matrix()

        # Crear emisiones simples
        np.random.seed(42)
        self.emissions = np.random.dirichlet(np.ones(self.n_vocab), size=self.n_states)

    def _normalize_matrix(self):
        """Normalizar matriz para que sea estoc√°stica"""
        row_sums = self.original.sum(axis=1, keepdims=True)
        row_sums[row_sums == 0] = 1
        self.original = self.original / row_sums
        print(f"‚úÖ Matriz normalizada: {self.n_states} estados")

    def method_1_node_shuffle(self, seed=42):
        """M√âTODO 1: Reordenar nodos (permutaci√≥n)"""
        print("üîÄ Aplicando permutaci√≥n de nodos...")

        np.random.seed(seed)
        permutation = np.random.permutation(self.n_states)

        # Aplicar permutaci√≥n
        self.hidden_matrix = self.original[permutation][:, permutation]
        self.key = {'method': 'shuffle', 'permutation': permutation}

        return self.hidden_matrix

    def method_2_noise_injection(self, noise_level=0.1):
        """M√âTODO 2: Inyectar ruido controlado"""
        print(f"üéõÔ∏è Inyectando ruido (nivel: {noise_level})...")

        # A√±adir ruido gaussiano
        noise = np.random.normal(0, noise_level, self.original.shape)
        noisy_matrix = self.original + noise

        # Asegurar valores positivos y normalizar
        noisy_matrix = np.abs(noisy_matrix)
        noisy_matrix = noisy_matrix / noisy_matrix.sum(axis=1, keepdims=True)

        self.hidden_matrix = noisy_matrix
        self.key = {'method': 'noise', 'noise_level': noise_level}

        return self.hidden_matrix

    def method_3_uniform_mixing(self, mix_ratio=0.3):
        """M√âTODO 3: Mezclar con matriz uniforme"""
        print(f"üåä Mezclando con matriz uniforme (ratio: {mix_ratio})...")

        # Crear matriz uniforme
        uniform_matrix = np.ones((self.n_states, self.n_states)) / self.n_states

        # Mezclar
        mixed_matrix = (1 - mix_ratio) * self.original + mix_ratio * uniform_matrix

        self.hidden_matrix = mixed_matrix
        self.key = {'method': 'uniform_mix', 'mix_ratio': mix_ratio}

        return self.hidden_matrix

    def method_4_state_duplication(self):
        """M√âTODO 4: Duplicar estados para confundir"""
        print("üë• Duplicando estados...")

        # Duplicar el primer estado
        n_new = self.n_states + 1
        expanded_matrix = np.zeros((n_new, n_new))

        # Copiar matriz original
        expanded_matrix[:self.n_states, :self.n_states] = self.original

        # Duplicar primer estado en la nueva posici√≥n
        expanded_matrix[self.n_states, :self.n_states] = self.original[0, :]
        expanded_matrix[:self.n_states, self.n_states] = self.original[:, 0]
        expanded_matrix[self.n_states, self.n_states] = self.original[0, 0]

        # Normalizar
        expanded_matrix = expanded_matrix / expanded_matrix.sum(axis=1, keepdims=True)

        self.hidden_matrix = expanded_matrix
        self.key = {'method': 'duplication', 'duplicated_state': 0}

        return self.hidden_matrix

    def method_5_cycle_shift(self, shift=1):
        """M√âTODO 5: Desplazamiento c√≠clico"""
        print(f"üîÑ Aplicando desplazamiento c√≠clico (shift: {shift})...")

        # Crear matriz de desplazamiento c√≠clico
        shifted_matrix = np.zeros_like(self.original)

        for i in range(self.n_states):
            for j in range(self.n_states):
                new_i = (i + shift) % self.n_states
                new_j = (j + shift) % self.n_states
                shifted_matrix[new_i, new_j] = self.original[i, j]

        self.hidden_matrix = shifted_matrix
        self.key = {'method': 'cycle_shift', 'shift': shift}

        return self.hidden_matrix

    def generate_sequences(self, matrix, n_sequences=200, seq_length=30):
        """Generar secuencias desde una matriz de transici√≥n"""
        sequences = []
        states_list = []

        for _ in range(n_sequences):
            # Estado inicial aleatorio
            current_state = np.random.choice(matrix.shape[0])

            sequence = []
            states = []

            for _ in range(seq_length):
                states.append(current_state)

                # Generar observaci√≥n
                obs = np.random.choice(self.n_vocab, p=self.emissions[current_state % self.n_states])
                sequence.append([obs])

                # Transici√≥n
                current_state = np.random.choice(matrix.shape[0], p=matrix[current_state])

            sequences.append(np.array(sequence))
            states_list.append(np.array(states))

        return sequences, states_list

    def train_hmm(self, sequences):
        """Entrenar HMM con las secuencias"""
        print("üß† Entrenando HMM...")

        # Preparar datos
        X_concat = np.concatenate(sequences)
        lengths = [len(seq) for seq in sequences]

        # Entrenar HMM
        model = hmm.MultinomialHMM(n_components=self.n_states, n_iter=100, random_state=42)
        model.fit(X_concat, lengths)

        self.learned_matrix = model.transmat_

        print(f"‚úÖ HMM entrenado")
        return model

    def recover_original(self):
        """Intentar recuperar matriz original usando la clave"""
        if self.key is None or self.learned_matrix is None:
            print("‚ùå No hay clave o matriz aprendida")
            return self.learned_matrix

        method = self.key['method']

        if method == 'shuffle':
            # Aplicar permutaci√≥n inversa
            perm = self.key['permutation']
            inv_perm = np.argsort(perm)
            recovered = self.learned_matrix[inv_perm][:, inv_perm]

        elif method == 'noise':
            # Para ruido, la matriz aprendida ya deber√≠a estar cerca
            recovered = self.learned_matrix

        elif method == 'uniform_mix':
            # Intentar "desmezclar" (aproximaci√≥n)
            mix_ratio = self.key['mix_ratio']
            uniform_matrix = np.ones((self.n_states, self.n_states)) / self.n_states
            recovered = (self.learned_matrix - mix_ratio * uniform_matrix) / (1 - mix_ratio)
            recovered = np.abs(recovered)
            recovered = recovered / recovered.sum(axis=1, keepdims=True)

        else:
            recovered = self.learned_matrix

        return recovered

    def visualize_comparison(self, title="Comparaci√≥n de Grafos"):
        """Visualizar los tres grafos: original, oculto, aprendido"""

        matrices = [
            (self.original, "üéØ Grafo Original"),
            (self.hidden_matrix, "üé≠ Grafo Oculto"),
            (self.learned_matrix, "üß† Grafo Aprendido")
        ]

        # A√±adir recuperado si existe
        if hasattr(self, 'recovered_matrix'):
            matrices.append((self.recovered_matrix, "üîÑ Grafo Recuperado"))

        n_plots = len(matrices)
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))
        axes = axes.flatten()

        for idx, (matrix, label) in enumerate(matrices):
            if idx >= 4:  # M√°ximo 4 subplots
                break

            ax = axes[idx]

            # Heatmap
            im = ax.imshow(matrix, cmap='Blues', vmin=0, vmax=1)
            ax.set_title(label, fontsize=12, fontweight='bold')
            ax.set_xlabel("Estado Destino")
            ax.set_ylabel("Estado Origen")

            # A√±adir valores en las celdas
            for i in range(matrix.shape[0]):
                for j in range(matrix.shape[1]):
                    text = ax.text(j, i, f'{matrix[i, j]:.2f}',
                                 ha="center", va="center", color="black", fontsize=8)

            # Colorbar
            plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

        # Ocultar subplots no usados
        for idx in range(n_plots, 4):
            axes[idx].set_visible(False)

        plt.suptitle(title, fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()

    def visualize_networks(self):
        """Visualizar como grafos de red"""

        matrices = [
            (self.original, "üéØ Original"),
            (self.hidden_matrix, "üé≠ Oculto"),
            (self.learned_matrix, "üß† Aprendido")
        ]

        fig, axes = plt.subplots(1, 3, figsize=(15, 5))

        for idx, (matrix, title) in enumerate(matrices):
            ax = axes[idx]

            # Crear grafo
            G = nx.DiGraph()

            # A√±adir nodos y aristas
            for i in range(matrix.shape[0]):
                for j in range(matrix.shape[1]):
                    if matrix[i, j] > 0.05:  # Solo mostrar conexiones significativas
                        G.add_edge(i, j, weight=matrix[i, j])

            # Layout
            pos = nx.spring_layout(G, seed=42)

            # Dibujar nodos
            nx.draw_networkx_nodes(G, pos, node_color='lightblue',
                                 node_size=800, ax=ax)

            # Dibujar aristas con grosor proporcional al peso
            edges = G.edges()
            weights = [G[u][v]['weight'] for u, v in edges]
            nx.draw_networkx_edges(G, pos, width=[w*5 for w in weights],
                                 edge_color='gray', arrows=True, ax=ax)

            # Etiquetas de nodos
            nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold', ax=ax)

            # Etiquetas de aristas (pesos)
            edge_labels = {(u, v): f'{w:.2f}' for u, v, w in G.edges(data='weight')}
            nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=8, ax=ax)

            ax.set_title(title, fontsize=12, fontweight='bold')
            ax.axis('off')

        plt.tight_layout()
        plt.show()

    def calculate_similarity(self, matrix1, matrix2):
        """Calcular similitud entre dos matrices"""
        # Correlaci√≥n
        corr = np.corrcoef(matrix1.flatten(), matrix2.flatten())[0, 1]

        # Error Frobenius
        frobenius_error = np.linalg.norm(matrix1 - matrix2, 'fro')

        # Error m√°ximo por entrada
        max_error = np.max(np.abs(matrix1 - matrix2))

        return {
            'correlation': corr,
            'frobenius_error': frobenius_error,
            'max_error': max_error
        }

    def analyze_hiding_effectiveness(self):
        """Analizar qu√© tan efectivo fue el m√©todo de ocultaci√≥n"""
        print("\nüìä AN√ÅLISIS DE EFECTIVIDAD DE OCULTACI√ìN")
        print("=" * 50)

        # Similitud Original vs Oculto (menor = mejor ocultaci√≥n)
        hide_sim = self.calculate_similarity(self.original, self.hidden_matrix)
        print(f"üé≠ Original vs Oculto:")
        print(f"   Correlaci√≥n: {hide_sim['correlation']:.4f} (menor = mejor)")
        print(f"   Error Frobenius: {hide_sim['frobenius_error']:.4f} (mayor = mejor)")

        # Similitud Original vs Aprendido (mayor = peor ocultaci√≥n)
        learn_sim = self.calculate_similarity(self.original, self.learned_matrix)
        print(f"\nüß† Original vs Aprendido:")
        print(f"   Correlaci√≥n: {learn_sim['correlation']:.4f}")
        print(f"   Error Frobenius: {learn_sim['frobenius_error']:.4f}")

        # Similitud Oculto vs Aprendido
        hidden_sim = self.calculate_similarity(self.hidden_matrix, self.learned_matrix)
        print(f"\nüîç Oculto vs Aprendido:")
        print(f"   Correlaci√≥n: {hidden_sim['correlation']:.4f}")
        print(f"   Error Frobenius: {hidden_sim['frobenius_error']:.4f}")

        # Score de ocultaci√≥n
        hiding_score = 1 - hide_sim['correlation']  # 1 = perfecta ocultaci√≥n
        detection_risk = learn_sim['correlation']   # 0 = no detectable

        print(f"\nüèÜ SCORES FINALES:")
        print(f"   üõ°Ô∏è Score de Ocultaci√≥n: {hiding_score:.4f} (0-1, mayor = mejor)")
        print(f"   ‚ö†Ô∏è Riesgo de Detecci√≥n: {detection_risk:.4f} (0-1, menor = mejor)")

        return {
            'hiding_score': hiding_score,
            'detection_risk': detection_risk,
            'similarities': {
                'original_vs_hidden': hide_sim,
                'original_vs_learned': learn_sim,
                'hidden_vs_learned': hidden_sim
            }
        }

def demo_all_methods():
    """Demostrar todos los m√©todos de ocultaci√≥n"""
    print("üé≠ DEMO: M√âTODOS DE OCULTACI√ìN EN CADENAS DE MARKOV")
    print("=" * 60)

    # Crear matriz original interesante
    original = np.array([
        [0.1, 0.7, 0.2],
        [0.4, 0.2, 0.4],
        [0.3, 0.3, 0.4]
    ])

    print("üéØ Matriz Original:")
    print(original)

    methods = [
        ('node_shuffle', 'method_1_node_shuffle', {}),
        ('noise_injection', 'method_2_noise_injection', {'noise_level': 0.15}),
        ('uniform_mixing', 'method_3_uniform_mixing', {'mix_ratio': 0.4}),
        ('cycle_shift', 'method_5_cycle_shift', {'shift': 1})
    ]

    results = {}

    for method_name, method_func, params in methods:
        print(f"\nüî¨ PROBANDO: {method_name.upper()}")
        print("-" * 40)

        # Crear instancia del sistema
        system = SimpleMarkovHiding(original)

        # Aplicar m√©todo de ocultaci√≥n
        getattr(system, method_func)(**params)

        # Generar datos desde la matriz oculta
        sequences, _ = system.generate_sequences(system.hidden_matrix)

        # Entrenar HMM
        model = system.train_hmm(sequences)

        # Intentar recuperar
        system.recovered_matrix = system.recover_original()

        # Analizar efectividad
        analysis = system.analyze_hiding_effectiveness()

        # Visualizar
        system.visualize_comparison(f"M√©todo: {method_name}")
        system.visualize_networks()

        results[method_name] = {
            'system': system,
            'analysis': analysis
        }

    # Comparaci√≥n final
    print(f"\nüèÜ COMPARACI√ìN FINAL DE M√âTODOS")
    print("=" * 50)

    for method_name, result in results.items():
        analysis = result['analysis']
        print(f"{method_name:15} | Ocultaci√≥n: {analysis['hiding_score']:.3f} | " +
              f"Detecci√≥n: {analysis['detection_risk']:.3f}")

    # Encontrar mejor m√©todo
    best_method = max(results.keys(),
                     key=lambda x: results[x]['analysis']['hiding_score'] -
                                  results[x]['analysis']['detection_risk'])

    print(f"\nü•á MEJOR M√âTODO: {best_method}")

    return results

# üöÄ EJECUTAR DEMO
if __name__ == "__main__":
    results = demo_all_methods()
```
